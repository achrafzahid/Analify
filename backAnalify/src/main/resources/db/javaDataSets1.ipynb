```python
import pandas as pd 
import numpy as np
```

```python
data =pd.read_csv("superstore_dataset.csv")
df=pd.DataFrame(data)
df.head()
```

```python
# construire region_id et le merger
df["region"] = (
    df["region"]
    .str.strip()
    .str.lower()
)

df_region = (
    df[["region"]]
    .drop_duplicates()
    .reset_index(drop=True)
)

df_region["region_id"] = df_region.index + 1
df = df.merge(df_region, on="region", how="left")
```

```python
#construire product_id
df["product_name"] = (
    df["product_name"]
    .str.strip()
    .str.lower()
)
df_product = (
    df[["product_name"]]
    .drop_duplicates()
    .reset_index(drop=True)
)

df_product["product_id"] = df_product.index + 1
df = df.merge(df_product, on="product_name", how="left")
```

```python
#construire category_id
df["category"] = (
    df["category"]
    .str.strip()
    .str.lower()
)
df_category = (
    df[["category"]]
    .drop_duplicates()
    .reset_index(drop=True)
)

df_category["category_id"] = df_category.index + 1
df = df.merge(df_category, on="category", how="left")
```

```python
# construire subcategorY_id
df["subcategory"] = (
    df["subcategory"]
    .str.strip()
    .str.lower()
)
df_subcategory = (
    df[["subcategory"]]
    .drop_duplicates()
    .reset_index(drop=True)
)

df_subcategory["subcategory_id"] = df_subcategory.index + 1
df = df.merge(df_subcategory, on="subcategory", how="left")
```

```python
#construire country_id
df["country"] = (
    df["country"]
    .str.strip()
    .str.lower()
)
df_country = (
    df[["country"]]
    .drop_duplicates()
    .reset_index(drop=True)
)

df_country["country_id"] = df_country.index + 1
df = df.merge(df_country, on="country", how="left")
```

```python
df["manufactory"] = (
    df["manufactory"]
    .str.strip()
    .str.lower()
)
df_manufactory = (
    df[["manufactory"]]
    .drop_duplicates()
    .reset_index(drop=True)
)

df_manufactory["manufactory_id"] = df_manufactory.index + 1
df = df.merge(df_manufactory, on="manufactory", how="left")
```

```python
df["state"] = (
    df["state"]
    .str.strip()
    .str.lower()
)
df_state = (
    df[["state"]]
    .drop_duplicates()
    .reset_index(drop=True)
)

df_state["state_id"] = df_state.index + 1
df = df.merge(df_state, on="state", how="left")
```

```python
df["store_id"] = (
    df["city"]
    .astype("category")
    .cat.codes + 1
)
```

```python
df.head()
```

```python
orders_unique = (
    df[["order_id", "zip"]]
    .drop_duplicates()
    .sort_values(["zip", "order_id"])
    .reset_index(drop=True)
)
orders_unique["zip_index"] = (
    orders_unique["zip"]
    .astype("category")
    .cat.codes
)
NB_CAISSIERS_PAR_ZIP = 2

orders_unique["caissier_id"] = (
    orders_unique.groupby("zip")
    .cumcount() % NB_CAISSIERS_PAR_ZIP
) + 1 + orders_unique["zip_index"] * NB_CAISSIERS_PAR_ZIP
df = df.merge(
    orders_unique[["order_id", "caissier_id"]],
    on="order_id",
    how="left"
)
df.head()
```

```python
orders_df=df[["order_id","order_date","ship_date","caissier_id"]]
orders_df["order_id"] = orders_df["order_id"].str.slice(start=8)
orders_df=orders_df.rename(columns={"caissier_id":"user_id"})
orders_df=orders_df.drop_duplicates()
orders_df.head()
```

```python
city_df=df[["zip","city","state_id"]]
city_df= (
    city_df
    .groupby("city", as_index=False)
    .agg({
        "zip": lambda x: x.mode().iloc[0]
    })
)

city_ids=city_df["zip"]

df_state = (
    df[df["zip"].isin(city_ids)]
    [["zip", "state_id"]]
    .drop_duplicates()
)

city_df = city_df.merge(
    df_state,
    left_on="zip",  # zip dans city_df
    right_on="zip",     # zip dans df
    how="left"
)

city_df=city_df.rename(columns={"zip":"city_id"})
```

```python
import pandas as pd

# Nombre de rangs (ex: 9 rangs = 3 catégories)
n_rangs = 9

rang_ids = []
rang_names = []
descriptions = []
category_ids = []

for rang_id in range(1, n_rangs + 1):
    # Chaque 3 rangs → même category_id
    category_id = ((rang_id - 1) // 3) + 1

    rang_ids.append(rang_id)
    rang_names.append(f"rang_{rang_id}")
    descriptions.append(
        f"This rank represents position level {rang_id} within the system hierarchy."
    )
    category_ids.append(category_id)

rang_df = pd.DataFrame({
    "rang_id": rang_ids,
    "rang_name": rang_names,
    "description": descriptions,
    "category_id": category_ids
})

rang_df.head()
```

```python
import pandas as pd

# Nombre total de faces
n_faces = 18

# Génération des données
face_ids = list(range(1, n_faces + 1))
face_names = [f"face_{i}" for i in face_ids]
rang_ids=[]

descriptions = []
for face_id in face_ids:
    rang_id = (face_id + 1) // 2  # chaque 2 faces → même rang
    descriptions.append(
        f"face {face_id} of the rang {rang_id}."
    )
    rang_ids.append(rang_id)

data = {
    "face_id": face_ids,
    "face_name": face_names,
    "description": descriptions,
    "rang_id":rang_ids
}

# Création du DataFrame
face_df = pd.DataFrame(data)

# Affichage
face_df.head()
```

```python
import pandas as pd
import random
from datetime import date, timedelta

# Paramètres
n_faces = 18
sections_per_face = 20
date_delai_value = date.today() + timedelta(days=30)

section_ids = []
section_names = []
face_ids = []
descriptions = []
base_prices = []
current_prices = []
statuses = []
date_delais = []

section_id = 1

for face_id in range(1, n_faces + 1):
    rang_id = (face_id + 1) // 2  # 2 faces par rang

    for position in range(1, sections_per_face + 1):
        # Prix multiple de 100 entre 2000 et 6000
        price = random.randrange(2000, 6001, 100)

        section_ids.append(section_id)
        section_names.append(f"section_{section_id}")
        face_ids.append(face_id)

        descriptions.append(
            f"Section {section_id} located on face_{face_id} of rang_{rang_id}, "
            f"position {position}."
        )

        base_prices.append(price)
        current_prices.append(price)  # égalité garantie
        statuses.append("CLOSE")
        date_delais.append(date_delai_value)

        section_id += 1

# Création du DataFrame
section_df = pd.DataFrame({
    "section_id": section_ids,
    "section_name": section_names,
    "face_id": face_ids,
    "description": descriptions,
    "base_price": base_prices,
    "current_price": current_prices,
    "date_delai": date_delais,
    "status": statuses
})

# Vérifications
section_df.head()

```
